##Data Insights and Visualizations

Top 10 highest-selling video games globally

```python
top_games = vg[['Name', 'Global_Sales']].groupby('Name').sum().nlargest(10, 'Global_Sales')
```

```python
#plt.figure(figsize=(6,4))
top_games['Global_Sales'].plot(kind='bar', color='skyblue')
plt.title("Top 10 Highest-Selling Video Games Globally")
plt.xlabel("Game Name")
plt.ylabel("Global Sales (in millions)")
plt.xticks(rotation=45, ha="right")
plt.show()
```

Most popular gaming platforms based on total global sales

```python
platform_sales = vg.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=True)

plt.figure(figsize=(7,6))
platform_sales.plot(kind='barh', color='blue')
plt.title("Total Global Sales by Platform")
plt.xlabel("Global Sales (in millions)")
plt.ylabel("Gaming Platform")
plt.show()
```

How has the number of video games released changed over the years?


```python
vg['Year_of_Release'] = pd.to_numeric(vg['Year_of_Release'], errors='coerce')
games_per_year = vg.groupby('Year_of_Release')['Name'].count()

plt.figure(figsize=(7,6))
plt.plot(games_per_year.index, games_per_year.values, marker='o', linestyle='-', color='red')
plt.title("Number of Games Released Over the Years")
plt.xlabel("Year of Release")
plt.ylabel("Number of Games")
plt.grid()
plt.show()
```

Most popular game genres based on global sales

```python
genre_sales = vg.groupby('Genre')['Global_Sales'].sum().sort_values()

plt.figure(figsize=(7,6))
genre_sales.plot(kind='barh', color='orange')
plt.title("Global Sales by Game Genre")
plt.xlabel("Total Global Sales (in millions)")
plt.ylabel("Genre")
plt.show()
```

Sales distribution across different regions

```python
region_sales = vg[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

plt.figure(figsize=(7,6))
region_sales.plot(kind='bar', color=['blue', 'green', 'red', 'purple'])
plt.title("Sales Distribution Across Different Regions")
plt.xlabel("Regions")
plt.ylabel("Total Sales (in millions)")
plt.show()
```

Which publishers have the highest global sales

```python
publisher_sales = vg.groupby('Publisher')['Global_Sales'].sum().nlargest(10)

plt.figure(figsize=(7,6))
publisher_sales.plot(kind='bar', color='cyan')
plt.title("Top 10 Publishers by Global Sales")
plt.xlabel("Publisher")
plt.ylabel("Global Sales (in millions)")
plt.xticks(rotation=45, ha="right")
plt.show()
```
